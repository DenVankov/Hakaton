```
def date(data):
  str_ = data.split('–')[1]
  str_ = str_.replace(' ', '')
  str_ = str_.split('.') 
  int_ =int(str_[2][:-1])*10000 + int(str_[1])*100 + int(str_[0])
  return int_
```
